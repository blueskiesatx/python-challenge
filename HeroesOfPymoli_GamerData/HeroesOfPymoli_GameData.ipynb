## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
### Top Spenders
#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#SN, Purchase Count, Average Purchase Price, and Total Purchase Value
top_spe = purchase_data[["SN","Price"]]
spend_group = top_spe.groupby("SN")
top_spe = spend_group.count()
#Formatting cells in price and value columns to currency
top_spe["Average Purchase Price"] = (spend_group["Price"].mean()).map("${:,.2f}".format)
top_spe["Total Purchase Value"] = (spend_group["Price"].sum())
top_spe = top_spe.rename(columns={"Price":"Purchase Count"})
top_spe = top_spe.sort_values(["Total Purchase Value"],ascending = False)
#Formatting cells in price and value columns to currency
top_spe["Total Purchase Value"] = top_spe["Total Purchase Value"].map("${:,.2f}".format)
#Printing dataframe for top 5 spenders' SN, purchase count, avg purchase price, and total purchase value
top_spe.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
### Most Popular Items
#Identify the 5 most popular items by purchase count, then list (in a table):
#Item ID, Item Name, Purchase Count, Item Price, and Total Purchase Value
most_pop = purchase_data[["Item ID","Item Name","Price"]]
mostpop_group = most_pop.groupby(["Item ID","Item Name"])
popular_item = mostpop_group.count()
#Sorting by price and purchase count
itprice = mostpop_group["Price"].first()
popular_item["Item Price"] = itprice
popular_item = popular_item.rename(columns={"Price":"Purchase Count"})
popular_item["Total Purchase Value"] = mostpop_group["Price"].sum()
#Formatting the price and value cells to currency after the maths
popular_item["Total Purchase Value"] = popular_item["Total Purchase Value"].map("${:,.2f}".format)
popular_item["Item Price"] = popular_item["Item Price"].map("${:,.2f}".format)
#Sorting the items by the purchase count
popular_item = popular_item.sort_values(["Purchase Count"],ascending = False)
#Printing dataframe
popular_item.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
### Most Profitable Items
#Identify the 5 most profitable items by total purchase value, then list (in a table):
#Item ID, Item Name, Purchase Count, Item Price, Total Purchase Value
most_profit = popular_item
most_profit["Total Purchase Value"] = most_profit["Total Purchase Value"].replace({'\$': '', ',': ''}, 
                                                                                        regex=True).astype(float)
most_profit = most_profit.sort_values(["Total Purchase Value"],ascending = False)
#Formatting the value cells to currency after the maths
most_profit["Total Purchase Value"] = most_profit["Total Purchase Value"].map("${:,.2f}".format)
#Printing dataframe
most_profit.head()
```
